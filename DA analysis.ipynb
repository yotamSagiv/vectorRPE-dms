```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = np.load('DA_trace_lowtemp.npz', allow_pickle=True)

for key in d.keys():
    print(key)
    
sl = d['simple_L']
sr = d['simple_R']
cl = d['chunk_L']
cr = d['chunk_R']
num_copies = d['num_copies']
thl = d['twohot_L']
thr = d['twohot_R']
fnl_reward_mags = d['reward_mags']
actions = d['actions']
trial_times = d['trial_times']
print(fnl_reward_mags)
# trial_times
```

### Simple encoding

```python
plt.figure(figsize=(12, 12))

# Plot left-preferring neurons
plt.subplot(2, 1, 1)
for i in range(9):
    plt.plot(cl[:, i], label='press %d' % i)

plt.xlim(2000, 2500)
plt.title('left-encoding DA neurons (n=27, 19 first-press neurons)')
plt.xlabel('timestep')
plt.ylabel('\"DA\" activity')
# plt.legend()
# plt.ylim(-6, 6)
plt.axhline(6 / num_copies, color='k', linestyle='--')
plt.axhline(-6 / num_copies, color='k', linestyle='--')


# plt.figure(figsize=(12, 12))

# Plot left-preferring neurons
plt.subplot(2, 1, 2)
for i in range(sl.shape[1]):
    plt.plot(sl[:, i], label='press %d' % i)

plt.xlim(2000, 2500)
plt.title('left-encoding DA neurons (n=9)')
plt.xlabel('timestep')
plt.ylabel('\"DA\" activity')
plt.ylim(-6, 6)

# Plot right-preferring neurons
# plt.subplot(2, 1, 2)
# for i in range(sr.shape[1]):
#     plt.plot(sr[:, i], label='press %d' % i)

# plt.xlim(2000, 2500)
# plt.legend()
# plt.title('right-encoding DA neurons')
# plt.xlabel('timestep')
# plt.ylabel('\"DA\" activity')
```

```python
plt.figure(figsize=(12, 6))
nts = sl.shape[0]

# Plot left-preferring neurons
# plt.subplot(2, 1, 1)
for i in range(sl.shape[1]):
    plt.plot(sl[:, i], label='press %d' % i)

plt.xlim(2275, 2300)
plt.title('left-encoding DA neurons')
plt.xlabel('timestep')
plt.ylabel('\"DA\" activity')
plt.scatter(np.arange(nts), actions, marker='x', color='k')

for i in range(len(trial_times)):
    plt.axvline(trial_times[i], color='k', linestyle='--')

# # Plot right-preferring neurons
# plt.subplot(2, 1, 2)
# for i in range(sr.shape[1]):
#     plt.plot(sr[:, i], label='press %d' % i)

# plt.xlim(2200, 2300)
# plt.legend()
# plt.title('right-encoding DA neurons')
# plt.xlabel('timestep')
# plt.ylabel('\"DA\" activity')
# plt.scatter(np.arange(nts), actions, marker='x', color='k')
```

```python
plt.figure(figsize=(12, 6))
nts = sl.shape[0]

# Plot left-preferring neurons
# plt.subplot(2, 1, 1)
plt.plot(np.sum(sl, axis=1), label='sum across neurons')
plt.legend()

plt.xlim(2275, 2300)
plt.title('left-encoding DA neurons, summed activity')
plt.xlabel('timestep')
plt.ylabel('\"DA\" activity')
plt.scatter(np.arange(nts), actions, marker='x', color='k')

for i in range(len(trial_times)):
    plt.axvline(trial_times[i], color='k', linestyle='--')

# # Plot right-preferring neurons
# plt.subplot(2, 1, 2)
# for i in range(sr.shape[1]):
#     plt.plot(sr[:, i], label='press %d' % i)

# plt.xlim(2200, 2300)
# plt.legend()
# plt.title('right-encoding DA neurons')
# plt.xlabel('timestep')
# plt.ylabel('\"DA\" activity')
# plt.scatter(np.arange(nts), actions, marker='x', color='k')
```
